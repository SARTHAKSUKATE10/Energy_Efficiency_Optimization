```python
def plot_correlation_matrix(df, name):
    numeric_columns=df.select_dtypes(include=[np.number])
    
    if numeric_columns.shape[1]==0:
        print(f"No numeric columns found in {name}, skipping correlation plot.")
        return
    
    correlation_matrix=numeric_columns.corr()
    
    plt.figure(figsize=(10,6))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
    plt.title(f"Correlation matrix of {name}")
    plt.show()
```

```python
plot_correlation_matrix(df1, "Union Processed Dataser")
plot_correlation_matrix(df2, "Sectorwise Merged Dataset")
```

```python
import missingno as msno
```

```python
def check_missing_values(df, name):
    missing_counts = df.isnull().sum()
    total_missing = missing_counts.sum()

    if total_missing == 0:
        print(f"No missing values found in {name}.")
    else:
        print(f"Missing values summary for {name}:\n")
        print(missing_counts[missing_counts > 0])
        
        # Visualize missing values
        plt.figure(figsize=(10, 6))
        msno.heatmap(df, cmap="viridis")
        plt.title(f"Missing Values Heatmap - {name}")
        plt.show()
```

```python
check_missing_values(df1, "Union Processed Dataset")
check_missing_values(df2, "Sectorwise Merged Dataset")
```

```python
missing_values = df1.isnull().sum() / len(df1) * 100
print("Missing values percentage in Union Processed Dataset:\n", missing_values)
```

```python
missing_values2 = df2.isnull().sum() / len(df2) * 100
print("\nMissing values percentage in Sectorwise Merged Dataset:\n", missing_values2)
```

```python
columns_to_fill = ["Temp (°C)", "Humidity (%)", "Solar Energy (kWh)", 
                   "Precipitation (mm)", "Season", "Population"]
```

```python
for col in columns_to_fill:
    missing_indices = df2[df2[col].isnull()].index  # Get indices where df2 has missing values
    df2.loc[missing_indices, col] = df1.loc[missing_indices, col]
```
