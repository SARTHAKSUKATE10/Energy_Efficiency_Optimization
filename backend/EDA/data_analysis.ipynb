```python
def plot_correlation_matrix(df, name):
    numeric_columns=df.select_dtypes(include=[np.number])
    
    if numeric_columns.shape[1]==0:
        print(f"No numeric columns found in {name}, skipping correlation plot.")
        return
    
    correlation_matrix=numeric_columns.corr()
    
    plt.figure(figsize=(10,6))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
    plt.title(f"Correlation matrix of {name}")
    plt.show()
```

```python
plot_correlation_matrix(df1, "Union Processed Dataser")
plot_correlation_matrix(df2, "Sectorwise Merged Dataset")
```

```python
def check_missing_values(df, name):
    missing_counts = df.isnull().sum()
    total_missing = missing_counts.sum()

    if total_missing == 0:
        print(f"No missing values found in {name}.")
    else:
        print(f"Missing values summary for {name}:\n")
        print(missing_counts[missing_counts > 0])
        
        # Visualize missing values
        plt.figure(figsize=(10, 6))
        msno.heatmap(df, cmap="viridis")
        plt.title(f"Missing Values Heatmap - {name}")
        plt.show()
```

```python
check_missing_values(df1, "Union Processed Dataset")
check_missing_values(df2, "Sectorwise Merged Dataset")
```

```python
df1['Date'] = pd.to_datetime(df1['Date'], errors='coerce')
df2['Date'] = pd.to_datetime(df2['Date'], errors='coerce')
```

```python
df1_numeric = df1.select_dtypes(include=[np.number])
df2_numeric = df2.select_dtypes(include=[np.number])
```

```python
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (data < lower_bound) | (data > upper_bound)
```

```python
outliers_df1_iqr = df1_numeric.apply(detect_outliers_iqr)
outliers_df2_iqr = df2_numeric.apply(detect_outliers_iqr)
```

```python
print("Outliers in df1 (IQR Method):")
print(outliers_df1_iqr.sum())
```

```python
print("\nOutliers in df2 (IQR Method):")
print(outliers_df2_iqr.sum())
```

```python
### Outlier Detection using Z-score
def detect_outliers_zscore(data, threshold=3):
    return np.abs(zscore(data)) > threshold
```

```python
outliers_df1_z = df1_numeric.apply(detect_outliers_zscore)
outliers_df2_z = df2_numeric.apply(detect_outliers_zscore)
```

```python
print("\nOutliers in df1 (Z-Score Method):")
print(outliers_df1_z.sum())
```

```python
print("\nOutliers in df2 (Z-Score Method):")
print(outliers_df2_z.sum())
```

```python
# Function to plot boxplots
def plot_boxplots(df, title):
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df)
    plt.title(f"Boxplot for {title}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
plot_boxplots(df1_numeric, "df1 (Union Processed Dataset)")
plot_boxplots(df2_numeric, "df2 (Sectorwise Modified Dataset)")
```
